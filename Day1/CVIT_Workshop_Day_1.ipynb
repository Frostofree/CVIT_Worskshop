### Problem 1

Given a list of binary strings `numbers`. Define a function that takes the first `k` strings and concats their parity into a string `encoding` and returns it.

````python
# Assumed that the problem was dealing wth columns, not individual integers in the list
def function(listx,k):
  encoding = ""
  a = k
  if k > len(listx):
    a = len(listx)

  for i in range(a):
    parity = 0
    for d in listx[i]:
      parity = parity ^ int(d)
    encoding += str(parity)

  return encoding



l = ["10001","111","000","111"]

print(function(l,4))
    

````

**Construct a frequency histogram for bell.jpg, and plot the histogram using plt.bar (search on google for syntax)**

````python
# The histogram Hf of the digital image f is a plot or graph of the frequency of occurrence of each gray level in f.

img_bell = cv2.cvtColor(cv2.imread("drive/MyDrive/images/images/bell.jpeg"),cv2.COLOR_BGR2RGB)
img_bell_gray = img_bell[:,:,0]*0.299 + img_bell[:,:,1]*0.587 + img_bell[:,:,2]*0.114
img_bell_gray = img_bell_gray.astype(np.int32).clip(0,255)
fig,axs = plt.subplots()
axs.imshow(img_bell_gray,cmap='gray')

# https://docs.opencv.org/3.4/d1/db7/tutorial_py_histogram_begins.html 

hist = cv2.calcHist([img_bell],[0],None,[256],[0,256])                # cv2.calcHist(images,channels,mask,jistSize,ranges[,hist[,accumulate]])

# print(hist)
# Above gives 256x1 array, but we also want BINS array to be generated as well, this is40x faster than the below implementation
# hist,bins = np.histogram(img_bell.ravel(),256,[0,256])

# print(hist.size)

f = plt.figure()          # Expand the histogram
f.set_figwidth(10)
f.set_figheight(5)

plt.subplot()
plt.plot(hist)
plt.xlim([0,255]) 
plt.show()

# plt.bar(hist,np.array([0,255]))
# plt.show()



````

**Perform thresholding on bell.jpg using the above constructed histogram.**

````python
# https://docs.opencv.org/4.x/d7/d4d/tutorial_py_thresholding.html

np.int32(img_bell_gray)

ret,thresh1 = cv2.threshold(img_bell,127,255,cv2.THRESH_BINARY)
ret,thresh2 = cv2.threshold(img_bell,127,255,cv2.THRESH_BINARY_INV)
ret,thresh3 = cv2.threshold(img_bell,127,255,cv2.THRESH_TRUNC)
ret,thresh4 = cv2.threshold(img_bell,127,255,cv2.THRESH_TOZERO)
ret,thresh5 = cv2.threshold(img_bell,127,255,cv2.THRESH_TOZERO_INV)
titles = ['Original Image','BINARY','BINARY_INV','TRUNC','TOZERO','TOZERO_INV']
images = [img_bell_gray, thresh1, thresh2, thresh3, thresh4, thresh5]
for i in range(6):
    plt.subplot(2,3,i+1),plt.imshow(images[i],'gray',vmin=0,vmax=255)
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
````

**Do a simple thresholding on the image sudoku.jpg, and find an apt threshold for the same using the above method.** 

````python
# global thresholding
img_sudoku = cv2.cvtColor(cv2.imread("drive/MyDrive/images/images/sudoku.jpg"),cv2.COLOR_BGR2RGB)

# print(img_sudoku.dtype)


ret,th1 = cv2.threshold(img_sudoku,65,255,cv2.THRESH_BINARY)            # Threhsold value is the second parameter
# plt.subplots()
# plt.imshow(thresh)


# Obviously not good, as one side is brighter while the other is dimmer. Thus we will use adaptive thresholding as well

th2 = cv2.adaptiveThreshold(img_sudoku,255,cv2.ADAPTIVE_THRESH_MEAN_C,cv2.THRESH_BINARY,11,2)               # Why error?
th3 = cv2.adaptiveThreshold(img_sudoku,255,cv2.ADAPTIVE_THRESH_GAUSSIAN_C,cv2.THRESH_BINARY,11,2)

titles = ['Original Image', 'Global Thresholding (v = 65)',
            'Adaptive Mean Thresholding', 'Adaptive Gaussian Thresholding']
images = [img_sudoku, th1, th2, th3]

for i in range(4):
    plt.subplot(2,2,i+1),plt.imshow(images[i],'gray')
    plt.title(titles[i])
    plt.xticks([]),plt.yticks([])
plt.show()
````
